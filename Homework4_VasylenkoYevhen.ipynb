### Завдання 2:

Потрібно проаналізувати всі товари на сайті: 
https://smallpacking.agrosem.ua/shop/

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```python
#Перевіряємо канал зв'язку
HEADERS = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1'
        }

SITE = 'https://smallpacking.agrosem.ua/shop/'
r = requests.get(SITE, HEADERS)
r.status_code
```

```python
#Шукаємо дані на сайті
soup = BeautifulSoup(r.text, 'lxml')
product = soup.find('div', class_="product-from-category-container")
product.span.text
```

```python
#Шукаємо внутрішні елементи
link = product.a.get('href')
link
```

```python
#Шукаємо ціну товару
def get_price(link):
    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    p = soup.find('span', class_="regular-price").text
    q = len(p)
    price = int(''.join(p[0:len(p)-4:1].split()[0:2]))    
    return price
```

```python
#Шукаємо код товару
def get_i_kod(link):
    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    i = soup.find('span', class_="sku").text
    i_kod = int(i[5:len(i)])    
    return i_kod
```

```python
#Шукаємо тип пакування
def get_pack(link):
    r = requests.get(link, HEADERS)
    soup = BeautifulSoup(r.text, 'lxml')
    pack = soup.find('span', class_="type-packaging-name").text    
    return pack
```

```python
#Створюємо посилання на інші сторінки сайту
for x in range(1, 11):
    link = f'https://smallpacking.agrosem.ua/shop/page/{x}/'
    print(link)
```

```python
#Налаштовуємо базу даних
connection = sqlite3.connect('products.db')
cursor = connection.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS product_list (Title TEXT, pack_type TEXT, KSU INT, Price INT)''')
connection.commit()
```

```python
#Завантажуємо дані з сайту
for i in range(1, 9):
    SITE = f'https://smallpacking.agrosem.ua/shop/page/{i}/'
    r = requests.get(SITE, HEADERS)
    if r.status_code == 200:
        soup = BeautifulSoup(r.text, 'lxml')
        product = soup.find('div', class_="product-from-category-container")
        for t in product.parent.find_all('div', class_="product-from-category-container"):
            name = t.span.text
            link = t.a.get('href')
            i_kod = get_i_kod(link)
            price = get_price(link)
            pack = get_pack(link)
            cursor.execute("INSERT INTO product_list VALUES (?,?,?,?)", (name, pack, i_kod, price))
            print(name, pack, i_kod, price)
```

```python
#Перевіярємо наповненість бази даних
for p in cursor.execute('SELECT * FROM product_list'):
    print(p)
```
